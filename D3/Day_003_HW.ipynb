```python
textPath = r'E:\Documents\2nd-ML100Days\D3\imagenet_urls_examples.txt'
f = open(textPath, encoding='utf8')
text = f.readlines()
imgName = []
url = []
for i in range(len(text)):
    imgName.append(text[i].split()[0])
    url.append(text[i].split()[1])
```

```python
labels = ['imgName', 'url']
cols = [imgName, url]
zipped = list(zip(labels, cols))
df = pd.DataFrame(dict(zipped))
df
```

```python
from urllib.request import urlretrieve
import os
os.chdir(r'E:\Documents\2nd-ML100Days\D3')

n = 0
c = 0
while c < 5:
    try:
        urlretrieve(url[n], imgName[n] + '.jpg')
        print('saved %s' % imgName[n])
        c += 1
        n += 1
    except:
        n += 1
        continue
```
