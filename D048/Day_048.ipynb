```python
from sklearn import datasets, metrics
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression, Lasso, Ridge
import pandas as pd
import numpy as np
import os
```

```python
trainPath = r'E:\Documents\2nd-ML100Days\D048'

train = pd.read_csv(os.path.join(trainPath, 'train.csv'), engine='python', encoding='utf8', header=None)
label = pd.read_csv(os.path.join(trainPath, 'trainLabels.csv'), engine='python', encoding='utf8', header=None)
test = pd.read_csv(os.path.join(trainPath, 'test.csv'), engine='python', encoding='utf8', header=None)
print(train.head())
print(label.head())
print(test.head())
```

```python
col_min = train.min(0)
col_max = train.max(0)
print(col_min.head(), col_max.head())
# normalize data
for i, col in enumerate(train.columns):
    train[col] = 2 * (train[col] - col_min[i])/(col_max[i] - col_min[i]) - 1
    test[col] = 2 * (test[col] - col_min[i])/(col_max[i] - col_min[i]) - 1
print(train.head(), test.head(), sep='\n')
```

```python
trainX, valX, trainY, valY = train_test_split(train, label, test_size=0.1)
print(trainX.shape, valX.shape, trainY.shape, valY.shape)
```

```python
clf = RandomForestClassifier()
clf.fit(trainX, trainY)
pred = clf.predict(valX)
acc = metrics.accuracy_score(valY, pred)
print(acc)
```

```python
result = clf.predict(test)
print(result, result.shape, sep='\n')
```

```python
resultPath = os.path.join(trainPath, 'result.csv')

cols = ['Id', 'Solution']
data = np.array([np.arange(test.shape[0]).reshape(-1,)+1, result]).T
print(data)
#print(test.shape)
res = pd.DataFrame(data, columns=cols)
print(res)
res.to_csv(resultPath, index=False)
```
